Cella 2.2: Caricamento delle immagini

```python
import shutil
import os
from google.colab import files

# Variabile per pulire la cartella relativa alle immagini caricate.
# Se si vogliono cancellare immagini caricate già in precedenza,
# impostare la variabile 'delete' a 'True'.
delete = False

# Se per qualche motivo, questa cella viene avviata all'interno della cartella
# principale del progetto, spostiamoci una cartella prima ("/content").
if(os.getcwd().endswith("InformativeDrawings")):
    examplesDirectory = "examples/personal/"
else:
    examplesDirectory = "InformativeDrawings/examples/personal/"

if (delete):
  if os.path.exists(examplesDirectory):
    shutil.rmtree(examplesDirectory)
  else:
    print("La cartella delle immagini non esiste.\nMolto probabilmente stai runnando la cella per la prima volta.\n")

# Se non esiste la cartella delle immagini, creiamola
if not os.path.exists(examplesDirectory):
    os.makedirs(examplesDirectory)

# Caricamento delle immagini
print("Caricamento immagini ...")
uploaded = files.upload()

# Variabile per salvataggio dei nomi delle immagini caricate
filenames = uploaded.keys()

# Per ogni immagine caricata, spostiamola nella cartella delle immagini
for f in filenames:
    shutil.move(f, examplesDirectory + f)

print("Caricamento immagini completato.")
```

# 3. Initialize the model and evaluate

Cella 3.1: Creazione e pulizia delle cartelle

```python
import gdown
import shutil
import os

# Se non esiste la cartella dei checkpoints, creiamola
if not os.path.exists("InformativeDrawings/checkpoints/"):
    os.makedirs("InformativeDrawings/checkpoints/")

# Download del modello e estrazione dello zip nella cartella apposita
gdown.download("https://drive.google.com/uc?id=11l5u5sb1PO5Z5YA3IoEHauVPm0k407C1&export=download&confirm=t", 'model.zip', quiet=False)
shutil.unpack_archive("model.zip", "InformativeDrawings/checkpoints/")

# Pulizia delle cartelle
print("Pulizia cartelle ...")
shutil.move("InformativeDrawings/checkpoints/model/anime_style/", "InformativeDrawings/checkpoints/anime_style/")
shutil.move("InformativeDrawings/checkpoints/model/contour_style/", "InformativeDrawings/checkpoints/contour_style/")
shutil.move("InformativeDrawings/checkpoints/model/opensketch_style/", "InformativeDrawings/checkpoints/opensketch_style/")
shutil.rmtree("InformativeDrawings/checkpoints/model/")
os.remove("model.zip")
print("Pulizia cartelle completata.")
```

# 4. Show results

```python
from matplotlib import pyplot as plt
import os
import cv2

# Scansione delle cartelle all'interno della cartella principale dei risultati
files = os.listdir("results")

# Per ogni cartella secondaria, analizziamone il contenuto e successivamente
# analizziamo e mostriamo ogni singola immagine presente in ogni cartella secondaria.
for f in files:
  l = os.listdir(os.path.join("results", f))
  for img in l:
    img_color = cv2.imread(os.path.join("results", f, img),1) # Lettura immagine
    plt.imshow(img_color) # Plot dell'immagine
    plt.title(img) # Titolo dell'immagine
    plt.axis("off") # Assi X e Y disabilitati
    plt.show() # Mostriamo l'immagine a video
```
